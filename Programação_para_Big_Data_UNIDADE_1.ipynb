# Variáveis
Agora que já aprendemos os tipos de dados em Python, vamos imergir em
como utilizá-los, bem como em como melhor utilizar os operadores aritméticos que
aprendemos.

Quando falamos de variável em programação, estamos falando de um espaço
na memória do computador destinado a um dado que é alterado durante a execução
de um programa.
Como o Python é uma linguagem orientada a objetos e tipagem dinâmica. Isto
significa que cada variável será um objeto e não é necessário declarar qual é o tipo de dados.


O processo de atribuição e declaração de variável em Python é com a utilização
do operador =. Ou seja, x=10, significa que a variável x será do tipo inteiro e recebe o valor 10. Vejamos mais alguns exemplos no quadro, com a explicação seguida de comentário

```
'''
Exemplo de Atribuição de variável
'''
x = 10 #Inicializando variável do tipo inteiro com valor 10
x = 20 #Atribuindo o valor 20
x = 17.5 #Atribuindo o valor 17.5
print(x) #impressao do valor de X
```

No exemplo acima qual é o valor de x? Se você já executou o programa, notou
que o valor de x foi a última atribuição, ou seja, x é 17.5, podendo perceber a atribuição dinâmica de uma variável.

# Operadores de Comparação

Os operadores de comparação são aqueles que comparam dois valores. Operadores de comparação são usados para comparar valores. Ele retorna True ou False de acordo com a condição.

• Operador Maior: operador a > b.


> Significado: o valor a é maior a que o valor b

```
print(20 > 50)
print(50>20)
print(20>20)
```

Operador Menor: operador a < b.


> Significado: o valor a é menor que o valor b?




```
print(20 < 50)
print(50<20)
print(20>20)
```

Operador Igual: operador a == b.
> Significado: o valor a é igual ao valor b?

```
print(20 == 50)
print(50==20)
print(20==20)
```

Operador Diferença: operador a != b.
> Significado: o valor a é diferente ao valor b?


```
print(20 != 50)
print(50 != 20)
print(20 != 20)
print(50 != 50)
```

• Operador Maior Igual: operador a >= b.
> Significado: o valor a é maior ou igual ao valor b?



```
print(20 >= 50)
print(50>=20)
print(20>=20)
```

• Operador Menor Igual: operador a <= b.
> Significado: o valor a é menor ou igual ao valor b?

```
print(20 >= 50)
print(50>=20)
print(20>=20)
```

Agora vamos combinar o conceito de operadores de comparação com a criação de variáveis.

```
x = 100
y = 120
print('A afirmação x > y é: ',x > y)
print('A afirmação x  < y é',x < y)
print('A afirmação x == y é',x==y)
print('A afirmação x != y é',x!=y)
print('A afirmação x >= y é',x>=y)
print('A afirmação x <= y é',x<=y)
```

# Operadores Lógicos
Operadores lógicos comparam valores booleanos, sejam armazenados em variáveis ou resultantes de operadores condicionais. Segundo Cechinel (2019), os operadores lógicos são usados para representar situações lógicas que não podem ser representadas por operadores aritméticos. Também são chamados conectivos lógicos por unirem duas expressões simples numa composta. Podem ser operadores binários, que operam em duas sentenças ou expressões, ou unário que opera numa sentença só.


```
x = True
y = False
print('O resultado da operação ( X AND Y)  é: ',x and y)
print('O resultado da operação ( X OR Y) é: ',x or y)
print('O resultado da operação ( NOT X) é:',not x)
```
